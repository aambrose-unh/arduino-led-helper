```python
import copy
import numpy as np
import matplotlib.pyplot as plt

# Define matrix class
class Matrix:
    def __init__(self, rows: int, cols: int, zigzag: bool = True):
        self.rows = rows
        self.cols = cols
        self.zigzag = zigzag

        self.matrix = [[(0, 0, 0) for _ in range(cols)] for _ in range(rows)]
        self.colors = {"red": (255, 0, 0),
                        "green": (0, 255, 0),
                        "blue": (0, 0, 255),
                        "brown": (150,75,0),
                        "yellow": (255, 255, 0),
                        "purple": (128, 0, 128),
                        "cyan": (0, 255, 255),
                        "orange": (255, 165, 0),
                        "pink": (255,192,203),
                        "black": (0, 0, 0),
                        "white": (255, 255, 255)}

    def __str__(self):
        return str(self.matrix)
    
    def __repr__(self):
        return str(self.matrix)

    @property
    def matrix(self):
        return self._matrix
    
    @matrix.setter
    def matrix(self, matrix):
        self._matrix = matrix
        self.rows = len(matrix)
        self.cols = len(matrix[0])
    
    # def get_indices(self):
    #     for row in range(1,self.rows+1):
    #         if self.zigzag and (row % 2 == 0):
    #             for col in range(1,self.cols+1):
    #                 yield row, (row * self.cols) - (col-1)
    #         else:
    #             for col in range(1,self.cols+1):
    #                 yield row, ((row-1) * self.cols) + col

    def get_indices(self):
        for row in range(self.rows):
            if self.zigzag and ((row+1) % 2 == 0):
                for col in range(self.cols):
                    yield row, col, ((row+1) * self.cols) - (col+1)
            else:
                for col in range(self.cols):
                    yield row, col, ((row) * self.cols) + col
    
    @property
    def colors(self):
        return self._colors

    @colors.setter
    def colors(self, colors):
        self._colors = colors

    def add_color(self, name, color):
        self.colors[name] = color

    def rgb_matrix(self):
        temp_matrix = copy.deepcopy(self.matrix)
        for row in range(len(temp_matrix)):
            for col in range(len(temp_matrix[row])):
                if temp_matrix[row][col] in self.colors:
                    temp_matrix[row][col] = self.colors[temp_matrix[row][col]]
        return temp_matrix
    
    def color_matrix(self):
        temp_matrix = copy.deepcopy(self.matrix)
        for row in range(len(temp_matrix)):
            for col in range(len(temp_matrix[row])):
                for color_name in self.colors:
                    if self.colors[color_name] == temp_matrix[row][col]:
                        temp_matrix[row][col] = color_name
        return temp_matrix
    
    def plot(self):

        # Plot the matrix as a color map
        plt.imshow(self.rgb_matrix())

        # Show the plot
        plt.show()
        
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a 3x3 matrix of RGB tuples using the 0 to 255 scale
rgb_matrix = np.array([
    [(255, 0, 0), (0, 255, 0), (0, 0, 255)],
    [(255, 255, 0), (0, 255, 255), (255, 0, 255)],
    [(0, 0, 0), (128, 128, 128), (255, 255, 255)]
]) / 255.0

# Define the size of the gaps between the points
gap_size = 1

# Create a new matrix with the gaps
new_matrix = np.zeros((int(rgb_matrix.shape[0] * (1 + gap_size) - gap_size)+1,
                       int(rgb_matrix.shape[1] * (1 + gap_size) - gap_size)+1,
                       rgb_matrix.shape[2]))
for i in range(rgb_matrix.shape[0]):
    for j in range(rgb_matrix.shape[1]):
        new_matrix[int(i * (1 + gap_size)):int(i * (1 + gap_size) + 1 + gap_size),
                   int(j * (1 + gap_size)):int(j * (1 + gap_size) + 1 + gap_size),
                   :] = rgb_matrix[i, j, :]

# Plot the new matrix as a color map
plt.imshow(new_matrix)

# Show the plot
plt.show()
```
